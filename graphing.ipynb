```python
import matplotlib
#matplotlib.use('Qt4Agg')
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from cosmoscatalog import Cosmos
import pandas
```

```python
cos = Cosmos()
cos.process()
rgb = cos.color()
```

```python
sns.set(style='white', palette='muted')
for a in xrange(4):
    print 'Aperture', str(a+1)
    g = sns.jointplot(rgb[a][0][0:1000], rgb[a][1][0:1000], kind="hex", size = 7, space=0, xlim=(0,3), ylim=(0,8))
    g.set_axis_labels('blue/green', 'red/green')
    plt.show()
```
